```python
import matplotlib.pyplot as plt
from skimage.io import imread_collection
import pandas as pd
import numpy as np
```

```python
mfinal=pd.read_csv('datafinal.csv')
mfinal=mfinal.drop(['Unnamed: 0'],axis=1)
mfinal
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

xp=mfinal[['RED','GREEN','BLUE','CONTRAST','ENTROPY','BRIGHT']]
yp=mfinal['PM']

x_train,x_test,y_train,y_test=train_test_split(xp,yp,test_size=0.4,random_state=10)

reg=LinearRegression()

reg.fit(x_train,y_train)

reg.predict(x_test)
reg.score(x_test,y_test)
```

```python
print(y_test,reg.predict(x_test))
```

```python
from sklearn.linear_model import LogisticRegression
reglog=LogisticRegression()
xp=mfinal[['RED','GREEN','BLUE','CONTRAST','ENTROPY','BRIGHT']]
yp=mfinal['Quality']
x_train,x_test,y_train,y_test=train_test_split(xp,yp,test_size=0.4,random_state=10)
reglog.fit(x_train,y_train)
reglog.predict(x_test)
reglog.score(x_test,y_test)
```

```python
y_predicted = reglog.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
df = pd.read_csv("datafinal.csv")
df
```

```python
inputs = df
```

```python
target = df['Quality']
```

```python
inputs_n = inputs.drop(['S.NO','Time','PM','Quality','Unnamed: 0'],axis='columns')
```

```python
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.4)

model = tree.DecisionTreeClassifier()

model.fit(x_train,y_train)

model.predict(x_test)
model.score(x_test,y_test)
```

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cd = confusion_matrix(y_test, y_predicted)
cd
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cd, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.ensemble import RandomForestClassifier
x_train,x_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.4)
rf=RandomForestClassifier(n_estimators=10)
rf.fit(x_train,y_train)

rf.predict(x_test)
rf.score(x_test,y_test)
```

```python
y_predicted = rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cr = confusion_matrix(y_test, y_predicted)
cr
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cr, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.naive_bayes import GaussianNB
x_train,x_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.3)
nb=GaussianNB()
nb.fit(x_train,y_train)

nb.predict(x_test)
nb.score(x_test,y_test)
```

```python
y_predicted = nb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cg = confusion_matrix(y_test, y_predicted)
cg
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cg, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.neighbors import KNeighborsClassifier
x_train,x_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.3)
knn=KNeighborsClassifier(n_neighbors=15)
knn.fit(x_train,y_train)

knn.predict(x_test)
knn.score(x_test,y_test)
```

```python
y_predicted = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
ck = confusion_matrix(y_test, y_predicted)
ck
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(ck, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.linear_model import SGDClassifier
x_train,x_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.3)
sgd=SGDClassifier(loss='modified_huber',shuffle=True,random_state=101)
sgd.fit(x_train,y_train)

sgd.predict(x_test)
sgd.score(x_test,y_test)
```

```python
y_predicted = sgd.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
csgd = confusion_matrix(y_test, y_predicted)
csgd
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(csgd, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
