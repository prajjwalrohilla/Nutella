```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("datacontrast.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_Quality = LabelEncoder()
```

```python

inputs = df
```

```python
inputs
```

```python
inputs['quality_n'] = le_Quality.fit_transform(inputs['Quality'])
```

```python
inputs
```

```python
target = df['quality_n']
```

```python
target
```

```python
inputs_n = inputs.drop(['S.NO','Time','PM','Quality','Unnamed: 0','quality_n'],axis='columns')
```

```python
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.5,random_state=10)

model = tree.DecisionTreeClassifier()

model.fit(x_train,y_train)

model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
