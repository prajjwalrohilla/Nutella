```python
import matplotlib.pyplot as plt
from skimage.io import imread_collection
import pandas as pd
import numpy as np
```

```python
mfinal=pd.read_csv('datafinal.csv')
mfinal=mfinal.drop(['Unnamed: 0'],axis=1)
mfinal
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

xp=mfinal[['RED','GREEN','BLUE','CONTRAST','ENTROPY']]
yp=mfinal['PM']

x_train,x_test,y_train,y_test=train_test_split(xp,yp,test_size=0.6,random_state=10)

reg=LinearRegression()

reg.fit(x_train,y_train)

reg.predict(x_test)
reg.score(x_test,y_test)
```

```python
print(x_test)
```

```python
print(y_test,reg.predict(x_test))
```

```python
from sklearn.linear_model import LogisticRegression
reglog=LogisticRegression()
xp=mfinal[['RED','GREEN','BLUE','CONTRAST','ENTROPY']]
yp=mfinal['Quality']
x_train,x_test,y_train,y_test=train_test_split(xp,yp,test_size=0.4,random_state=10)
reglog.fit(x_train,y_train)
reglog.predict(x_test)
reglog.score(x_test,y_test)
```

```python

y_predicted = reglog.predict(x_test)
```

```python
print(y_predicted)
```

```python
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python

import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.svm import SVC
model= SVC()
xp=mfinal[['RED','GREEN','BLUE','CONTRAST','ENTROPY']]
yp=mfinal['Quality']
x_train,x_test,y_train,y_test=train_test_split(xp,yp,test_size=0.4)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cs = confusion_matrix(y_test, y_predicted)
cs
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cs, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
